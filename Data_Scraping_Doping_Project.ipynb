```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# GET request to the page
url = "https://www.usada.org/news/sanctions/"
response = requests.get(url)
```

```python
# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Find the table and extract rows
table = soup.find('table')
rows = table.find_all('tr')
```

```python
# Extract data from the rows
data = []
for row in rows[1:]:  # Skip the header row
    cells = row.find_all('td')
    athlete = cells[0].get_text(strip=True)
    sport = cells[1].get_text(strip=True)
    substance = cells[2].get_text(strip=True)
    sanction_terms = cells[3].get_text(strip=True)
    date = cells[4].get_text(strip=True)
    data.append([athlete, sport, substance, sanction_terms, date])
```

```python
# Create a DataFrame, rename the columns and save to CSV
columns = ['athlete', 'sport', 'substance_or_reason', 'sanction_terms', 'sanction_announced']
df = pd.DataFrame(data, columns=columns)
```

```python
# Save to CSV
df.to_csv('usada_sanctions.csv', index=False)
```
